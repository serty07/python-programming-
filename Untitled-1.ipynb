# Introduction to Python

"Python is an interpreted, object-oriented, high-level programming language with dynamic semantics. Its high-level built in data structures, combined with dynamic typing and dynamic binding, make it very attractive for Rapid Application Development, as well as for use as a scripting or glue language to connect existing components together. Python’s simple, easy to learn syntax emphasizes readability and therefore reduces the cost of program maintenance. Python supports modules and packages, which encourages program modularity and code reuse. The Python interpreter and the extensive standard library are available in source or binary form without charge for all major platforms, and can be freely distributed."

# What is interactive programming in python? -(5 Marks)


1. When we can run a programing direly from terminal , and the output is instant. We dont need a helper command to run the script.
2. Interactive programming in Python means writing and executing Python commands directly in the Python interpreter (terminal) and getting the output instantly.
3. In this mode, we don’t need to save the code in a file and we don’t need to run it using a script command every time.
4. It is mainly done in the Python Shell/REPL, where Python takes input line by line and shows output immediately.

# What is script programming in python? - (5 Marks)

1. When we submitt the script to the compiler using a helper command. In the case of python the terminal command is "python - file name"
2. Script programming in Python means writing Python code in a file (example: program.py) and then running the complete program using a command in the terminal.
3. In this mode, the program is saved as a script, and Python executes it line by line when we run it.
